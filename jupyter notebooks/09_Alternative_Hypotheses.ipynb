# Alternative Hypotheses: Main Ideas!!!

In the field of statistics, the **null hypothesis** is a general statement or default position that there is no relationship between two measured phenomena. 

The **alternative hypothesis** is the hypothesis used in hypothesis testing that is contrary to the null hypothesis. It is usually taken to be that the observations are the result of a real effect (with some amount of chance variation superposed). 

In this notebook, we are going to discuss about the alternative hypotheses, its main ideas and how it is used in statistical tests.

```python

# Visualizing Null and Alternative Hypotheses
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-4, 4, 1000)
null_hypothesis = np.exp(-x**2)  # Example null hypothesis curve
alt_hypothesis = np.exp(-((x - 2)**2) / 2)  # Example alternative hypothesis curve

plt.plot(x, null_hypothesis, label='Null Hypothesis (H0)', color='blue')
plt.plot(x, alt_hypothesis, label='Alternative Hypothesis (H1)', color='orange')
plt.title("Null vs. Alternative Hypotheses")
plt.xlabel("Test Statistic")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()
```

## Hypothesis Testing

```python

# Simulated Flowchart of Hypothesis Testing (Text-based Example)
print("Hypothesis Testing Steps:")
print("1. Define H0 and H1")
print("2. Choose significance level (α)")
print("3. Collect data and compute test statistic")
print("4. Compare p-value to α")
print("5. Make a decision: Reject or fail to reject H0")
```

Hypothesis testing is a statistical method that is used in making statistical decisions using experimental data. Hypothesis Testing is basically an assumption that we make about the population parameter.

Hypothesis testing involves the careful construction of two statements: the null hypothesis and the alternative hypothesis.

These hypotheses contain opposing viewpoints. 

- **H0**: The null hypothesis: It is a statement about the population that either is believed to be true or is used to put forth an argument unless it can be shown to be incorrect beyond a reasonable doubt. 

- **H1**: The alternative hypothesis: It is a claim about the population that is contradictory to H0 and what we conclude when we reject H0.

```python

# Visualizing p-values with an example
from scipy.stats import norm

# Define a distribution
x = np.linspace(-100, 100, 1000000)
pdf = norm.pdf(x)
pdf /= np.sum(pdf)
critical_value = 1.96  # 5% significance level for a two-tailed test

plt.plot(x, pdf, label="Normal Distribution")
plt.fill_between(x, 0, pdf, where=(x > critical_value) | (x < -critical_value), color="red", alpha=0.5, label="Rejection Region")
plt.title("P-value and Rejection Regions")
plt.xlabel("Test Statistic")
plt.ylabel("Probability Density")
plt.xlim([-4,4])
plt.legend()
plt.grid(True)
plt.show()
```

The alternative hypothesis plays a crucial role in hypothesis testing. It's what we're testing, to see if our null hypothesis is not true. A statistical test needs three things: 

1. **Data**: The information we have collected and wish to analyse. 
2. **Null Hypothesis (H0)**: A statement of no effect that we can test against our alternative hypothesis. 
3. **Alternative Hypothesis (H1)**: The statement we wish to test for.

```python

# Simulating hypothesis test with two groups
import numpy as np
from scipy.stats import ttest_ind

# Simulate two groups of data
group1 = np.random.normal(10, 2, 30)  # Mean = 10, SD = 2
group2 = np.random.normal(12, 2, 30)  # Mean = 12, SD = 2

# Perform t-test
t_stat, p_value = ttest_ind(group1, group2)

# Display results
print(f"T-statistic: {t_stat:.2f}, P-value: {p_value:.4f}")

# Visualize data
import matplotlib.pyplot as plt

plt.boxplot([group1, group2], labels=["Group 1", "Group 2"])
plt.title("Comparison of Two Groups")
plt.ylabel("Values")
plt.grid(True)
plt.show()
```

When we only have two groups of data, the alternative hypothesis is super obvious because it is just the opposite of the null hypothesis. However, when we have three or more groups, we have options for the alternative hypothesis. Depending on which one we use in the statistical test, we can end up making a different decision about the null hypothesis. 

It's important to clearly state which alternative hypothesis we want to use. However, regardless of the alternative hypothesis we used in the test, we only reject or fail to reject the primary or null hypothesis.
