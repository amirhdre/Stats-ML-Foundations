# Bar Charts Are Better Than Pie Charts

In data visualization, selecting the right chart depends on the story you want to tell. Pie charts and bar charts are common choices, but bar charts often outperform pie charts in clarity and ease of interpretation.

This notebook demonstrates why bar charts are generally better than pie charts using Python's `matplotlib` library.

## Why Compare Pie and Bar Charts?

Pie charts display data as proportions of a whole but struggle with precise comparisons. Bar charts, on the other hand, excel at showing differences clearly.

Consider a dataset showing how Herman spends his day across four activities: sleeping, working, eating, and resting.

```python
# Import necessary libraries
import matplotlib.pyplot as plt
import numpy as np

# Data for Herman's daily activities
activities = ['Sleeping', 'Working', 'Eating', 'Resting']
time_spent = [8, 9, 2, 5]  # Hours spent on each activity
```

### Visualizing with Pie and Bar Charts

First, let's visualize Herman's daily activities using a pie chart and a bar chart.

```python
# Pie chart
plt.figure(figsize=(5, 5))
plt.pie(time_spent, labels=activities, autopct='%1.1f%%', colors=['lightblue', 'lightgreen', 'pink', 'orange'])
plt.title('Pie Chart: Herman\'s Daily Activities')
plt.show()

# Bar chart
plt.figure(figsize=(8, 5))
plt.bar(activities, time_spent, color=['lightblue', 'lightgreen', 'pink', 'orange'])
plt.xlabel('Activities')
plt.ylabel('Time Spent (Hours)')
plt.title('Bar Chart: Herman\'s Daily Activities')
plt.show()
```

### Comparing Insights

The pie chart shows proportions but makes it difficult to compare specific values. In contrast, the bar chart clearly displays the time Herman spends on each activity, making comparisons straightforward.

## Comparing Multiple Data Sets

Now, let's compare the daily activities of Herman and Earl. Pie charts struggle to represent multiple datasets, whereas bar charts excel.

```python
# Earl's daily activities data
time_spent_earl = [7.5, 8.5, 3, 6]

# Side-by-side pie charts
fig, ax = plt.subplots(1, 2, figsize=(12, 6))
ax[0].pie(time_spent, labels=activities, autopct='%1.1f%%', colors=['lightblue', 'lightgreen', 'pink', 'orange'])
ax[0].set_title('Pie Chart: Herman\'s Activities')
ax[1].pie(time_spent_earl, labels=activities, autopct='%1.1f%%', colors=['lightblue', 'lightgreen', 'pink', 'orange'])
ax[1].set_title('Pie Chart: Earl\'s Activities')
plt.show()

# Bar chart comparison
barWidth = 0.35
r1 = np.arange(len(time_spent))
r2 = [x + barWidth for x in r1]

plt.figure(figsize=(10, 5))
plt.bar(r1, time_spent, color='lightblue', width=barWidth, edgecolor='grey', label='Herman')
plt.bar(r2, time_spent_earl, color='lightgreen', width=barWidth, edgecolor='grey', label='Earl')
plt.xlabel('Activities')
plt.ylabel('Time Spent (Hours)')
plt.xticks([r + barWidth / 2 for r in range(len(time_spent))], activities)
plt.title('Bar Chart: Herman\'s vs Earl\'s Activities')
plt.legend()
plt.show()
```

### Insights from the Comparison

- Pie charts make it hard to compare Herman's and Earl's activities.
- Bar charts clearly show:
  - Herman works slightly more than Earl.
  - Earl eats and rests more than Herman.
This demonstrates why bar charts are more effective for comparing multiple datasets.

## Conclusion

While pie charts are visually appealing, they are often harder to interpret, especially when comparing datasets. Bar charts provide a clear, accurate, and space-efficient alternative. Always choose the visualization that best communicates your data's story.
