# The Main Ideas behind Probability Distributions

Let's consider that we have measured the height of a lot of people. The measurements we made ranged from 5.1 feet to 6.3 feet. If we categorize these measurements into bins, for instance, a bin that spans from 5 feet to 5.5 feet, and continue this process for all the measurements, we obtain a histogram. A histogram is a graphical representation of the distribution of a dataset. In this case, the histogram gives us a sense of how likely we'll measure someone really tall or really short or closer to the average.

Most of the measurements come from people between 5 and 6 feet tall, while people smaller than 5 feet tall or taller than 6 feet tall were relatively rare. In other words, if you picked one measurement at random, there's a good chance it would be between 5 and 6 feet tall. This gives us an idea of the distribution of the heights of people.

What if we use smaller bin sizes for our measurements? Now the bins are half as wide as before. By measuring more people and using smaller bins, we get a more accurate and more precise estimate of how heights are distributed.

We can also use a curve to approximate the histogram. The curve tells us the same thing that the histogram tells us. However, the curve has a few advantages over the histogram:

- Even though we measured a bunch of people, we didn't get a value for certain bins. While we can't calculate this probability with the histogram, we can use the curve to calculate the probability.
- The curve is not limited by the width of the bins. If we wanted to know the probability of measuring someone between any two specific values, we could use calculus, or a computer to calculate this, without having to round to the nearest bin size.
- If we don't have enough time or money to get a lot of measurements, the approximate curve based on the mean and standard deviation of the data that we were able to collect is usually good enough. Thus, using the curve can save us a lot of time and money.

Both the histogram and the curve are distributions. They show us how the probabilities of measurements are distributed. The tallest part of the histogram, or curve, shows the region where measurements are most likely. The low parts of the histogram, or curve, show where measurements are less likely.

We've been talking about how height measurements are distributed, but there are all kinds of distributions with all kinds of interesting shapes which we will explore in the future.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate synthetic height data
np.random.seed(42)  # For reproducibility
mean_height = 5.8  # Average height in feet
std_dev = 0.2  # Standard deviation in feet
num_samples = 1000  # Number of samples

# Simulate heights using a normal distribution
heights = np.random.normal(mean_height, std_dev, num_samples)

# Create histogram
bin_width = 0.1  # Width of each bin
bins = np.arange(5.0, 6.5, bin_width)  # Bin edges

plt.figure(figsize=(20,8))

plt.hist(heights, bins=bins, density=True, alpha=0.6, color='skyblue', label='Histogram')

# Fit a normal distribution to the data
mu, sigma = norm.fit(heights)  # Mean and standard deviation

# Generate the PDF curve
x = np.linspace(5.0, 6.5, 500)  # Values for the curve
pdf = norm.pdf(x, mu, sigma)

plt.plot(x, pdf, 'r-', label=f'Normal PDF\n($\mu={mu:.2f}$, $\sigma={sigma:.2f}$)')

# Add labels and legend
plt.title('Height Distribution')
plt.xlabel('Height (feet)')
plt.ylabel('Probability Density')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
