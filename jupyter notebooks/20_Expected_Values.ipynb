# Expected Value

Expected value is a concept used in statistics to predict the long-term behavior of random events. It represents an average amount of some quantity per observation, over an arbitrarily large number of observations.

Expected value is often used to decide whether a certain action is advantageous or not, based on its probable outcomes. It is particularly useful in situations where an action can be repeated many times, because the law of large numbers ensures that the average of the results obtained from a large number of trials should be close to the expected value.

Now, let's dive into the concept of expected value using a case from a place called Statland.

## Case: Random Bet in Statland

There are 213 people living in Statland, out of which 37 people have heard of the movie 'Troll 2'. The rest, 176 people, have never heard of it. Now, a friend named Statsquatch wants to make a bet. He says, "I bet you one dollar that the next person we meet has heard of the movie 'Troll 2'."

We can calculate the expected value of this bet to decide whether to accept the bet or not.

### Welcome to Statland

Statland is a small town with 213 people. To understand probabilities and expected values, imagine walking through the town and making bets about random events, like whether the next person you meet has heard of the movie 'Troll 2'. Below is a visualization of the town's population split based on this scenario.

```python
import matplotlib.pyplot as plt

# Population data
labels = ['Heard of "Troll 2"', 'Not Heard of "Troll 2"']
sizes = [37, 176]
colors = ['skyblue', 'lightcoral']
explode = (0.1, 0)  # explode the first slice

# Pie chart
plt.figure(figsize=(6, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title("Population Distribution in Statland")
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

## Calculation of Probabilities

First, let's calculate the probabilities of meeting a person in Statland that has heard and hasn't heard of 'Troll 2'.

The probability of a person having heard of 'Troll 2' is calculated by dividing the number of people who have heard of 'Troll 2' by the total population of Statland. The same goes for the probability of a person not having heard of 'Troll 2'.

```python
# Calculation of Probabilities

total_population = 213
heard_troll2 = 37
not_heard_troll2 = 176

probability_heard = heard_troll2 / total_population
probability_not_heard = not_heard_troll2 / total_population

probability_heard, probability_not_heard
```

### Probability Visualization

Probabilities help us understand how likely each outcome is. Below, we visualize the probability of meeting someone who has heard of 'Troll 2' versus someone who hasn't.

```python
# Visualization of probabilities
probabilities = [37 / 213, 176 / 213]
categories = ['Heard of "Troll 2"', 'Not Heard of "Troll 2"']

plt.figure(figsize=(8, 4))
plt.bar(categories, probabilities, color=['skyblue', 'lightcoral'], alpha=0.7)
plt.title("Probability of Meeting Someone in Statland")
plt.ylabel("Probability")
plt.ylim(0, 1)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

### Expected Value Formula

The expected value can be calculated using the formula:

$$EV = \sum_{i} (P(i) \times Outcome(i))$$

For our scenario:
- If the next person **has heard** of 'Troll 2', we lose 1 dollar.
- If the next person **hasn't heard**, we gain 1 dollars.

Below is a calculation of the expected value.

The calculated probabilities indicate that the probability of meeting a person who has heard of 'Troll 2' is relatively low (0.17), while the probability of meeting a person who hasn't heard of 'Troll 2' is relatively high (0.83).

Let's use these probabilities to calculate the expected value for the bet.

## Calculation of Expected Value

The expected value of a bet is calculated by summing the products of each outcome and its associated probability. In this case, there are two outcomes: losing one dollar if the next person has heard of 'Troll 2' (with a probability of 0.17), and winning one dollar if the next person hasn't heard of 'Troll 2' (with a probability of 0.83).

```python
# Calculation of Expected Value

outcome_heard = -1  # losing one dollar
outcome_not_heard = 1  # winning one dollar

expected_value = (probability_heard * outcome_heard) + (probability_not_heard * outcome_not_heard)
expected_value
```

The calculated expected value is 0.66. This means that on average, we expect to gain 66 cents every time we make this bet. Thus, if we can make this bet a bunch of times, we should make money in the long run, even though we will lose some of the time.

## Expected Value with Different Outcomes

Now, imagine Statsquatch changing the conditions of the bet. He says, "Because it is relatively rare for someone in Statland to have heard of 'Troll 2', I will pay you 10 dollars if the next person we meet has heard of 'Troll 2'. But if they have not, you pay me 1 dollars."

Let's calculate the expected value for this new bet.

```python
# Calculation of Expected Value for the New Bet

outcome_heard = 10  # winning ten dollars
outcome_not_heard = -1  # losing one dollar

expected_value = (probability_heard * outcome_heard) + (probability_not_heard * outcome_not_heard)
expected_value
```

The calculated expected value for the new bet is 0.87. This means we expect to gain on average 87 cents every time we make this new bet.

Thus, even though the outcome for a person having heard of 'Troll 2' is now much more profitable ($10), the low probability of this event (0.17) compensates for the high gain, resulting in a similar expected value as the previous bet.
