# Pearson's Correlation

Correlation is a statistical measure that describes the association between random variables. In this notebook, we will provide a detailed explanation of Pearson's Correlation, one of the most widely used measures of correlation.

The Pearson correlation, also called Pearson's r, is a measure that quantifies the strength and direction of a relationship between two variables. It is named after the British statistician Karl Pearson. The Pearson correlation is a parametric measure. This means that it makes certain assumptions about the data, which we will discuss later.

The Pearson correlation is used when both variables are normally distributed and there is a linear relationship between the variables. In this case, the Pearson correlation coefficient can provide an accurate measure of the relationship. The Pearson correlation coefficient is a value between -1 and 1, where 1 means a perfect positive linear relationship, -1 means a perfect negative linear relationship, and 0 means no linear relationship.

Let's start by importing the necessary libraries and creating some data.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Create two variables with a linear relationship
x = np.linspace(-5, 5, 100)
y = 3 * x + np.random.normal(0, 2, 100)

# Plot the variables
plt.scatter(x, y)
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.title('Scatterplot of two variables with a linear relationship')
plt.show()
```

```python
# Example visualization of Pearson's correlation
import numpy as np
import matplotlib.pyplot as plt

# Generate example data
np.random.seed(42)
x = np.random.rand(100)
y = 2 * x + np.random.normal(0, 0.1, 100)

# Scatter plot with trend line
plt.figure(figsize=(8, 5))
plt.scatter(x, y, label='Data Points', color='blue', alpha=0.7)
plt.plot(np.unique(x), np.poly1d(np.polyfit(x, y, 1))(np.unique(x)), color='red', label='Trend Line')
plt.title("Scatter Plot with Pearson Correlation")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

We can see that there is a linear relationship between the two variables. The Pearson correlation coefficient can be calculated using the `pearsonr` function from the `scipy.stats` module. This function returns two values: the correlation coefficient and the p-value for testing non-correlation.

```python
# Calculate Pearson correlation coefficient and p-value
r, p = stats.pearsonr(x, y)
print(f'Pearson correlation coefficient: {r:.2f}')
print(f'p-value: {p:.2f}')
```

The Pearson correlation coefficient is close to 1, which indicates a strong positive linear relationship. The p-value is very small, which means that the probability of observing such a strong linear relationship by chance is extremely low.

Now let's look at two variables with no relationship.

```python
# Create two variables with no relationship
x = np.linspace(-5, 5, 100)
y = np.random.normal(0, 2, 100)

# Plot the variables
plt.scatter(x, y)
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.title('Scatterplot of two variables with no relationship')
plt.show()
```

```python

# Scatter plot with trend line
plt.figure(figsize=(8, 5))
plt.scatter(x, y, label='Data Points', color='blue', alpha=0.7)
plt.plot(np.unique(x), np.poly1d(np.polyfit(x, y, 1))(np.unique(x)), color='red', label='Trend Line')
plt.title("Scatter Plot with Pearson Correlation")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

We can see that there is no apparent relationship between the two variables. Let's calculate the Pearson correlation coefficient and the p-value.

```python
# Calculate Pearson correlation coefficient and p-value
r, p = stats.pearsonr(x, y)
print(f'Pearson correlation coefficient: {r:.2f}')
print(f'p-value: {p:.2f}')
```

## Assumptions of Pearson's Correlation

The Pearson correlation coefficient is a parametric measure, which means it makes certain assumptions about the data. These are:

1. **Linearity**: The relationship between the two variables is linear. This means that the variables increase or decrease at a constant rate.
2. **Homoscedasticity**: The variance of one variable is stable across the range of the other variable. This means that the spread of values is consistent across the variables.
3. **Independence**: The observations are independent of each other. This means that the value of one observation does not influence or is influenced by the value of another observation.
4. **Normality**: Both variables are normally distributed. However, this assumption can be relaxed if the sample size is large.

If these assumptions are not met, the Pearson correlation coefficient can be misleading. In these cases, it is better to use a non-parametric measure of correlation, such as Spearman's rank correlation.
