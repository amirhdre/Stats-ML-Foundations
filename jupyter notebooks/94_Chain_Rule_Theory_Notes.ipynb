
# Chain Rule - Theory Notes

## Introduction
The chain rule is a fundamental concept in calculus used to compute the derivative of a composite function. If $z$ depends on $y$, and $y$ depends on $x$, then $z$ depends on $x$, and the chain rule states:

\[
\frac{dz}{dx} = \frac{dz}{dy} \cdot \frac{dy}{dx}
\]

## Quick Recap on Derivatives
The derivative represents the rate of change of a function. For example:

If $f(x) = x^2$, then $f'(x) = 2x$.

## Chain Rule Explained
If we have a composition of functions $f(g(x))$, the derivative is given by:

\[
\frac{d}{dx} f(g(x)) = f'(g(x)) \cdot g'(x)
\]

### Example
1. Let $y(x) = (3x^2 + 2)^5$. By the chain rule:
   \[
   \frac{dy}{dx} = 5(3x^2 + 2)^4 \cdot 6x = 30x(3x^2 + 2)^4
   \]

2. Suppose $h(w) = 2w$ and $s(h) = \frac{1}{4}h$. Then:
   \[
   \frac{ds}{dw} = \frac{ds}{dh} \cdot \frac{dh}{dw} = \frac{1}{4} \cdot 2 = \frac{1}{2}
   \]

## Steps to Apply the Chain Rule
1. Identify the outer and inner functions.
2. Differentiate the outer function while keeping the inner function unchanged.
3. Multiply by the derivative of the inner function.

## Common Scenarios
1. **Exponential Functions**:
   \[
   y = e^{x^2} \implies \frac{dy}{dx} = e^{x^2} \cdot 2x
   \]

2. **Composite Functions**:
   \[
   y = \sqrt{3x + 1} \implies \frac{dy}{dx} = \frac{1}{2\sqrt{3x + 1}} \cdot 3
   \]

3. **Physical Relationships**:
   - If $T(t) = 5t^2$ and $t(x) = 3x + 1$:
     \[
     \frac{dT}{dx} = \frac{dT}{dt} \cdot \frac{dt}{dx} = 10t \cdot 3 = 30(3x + 1)
     \]

These principles are fundamental for solving real-world problems and understanding mathematical modeling.
