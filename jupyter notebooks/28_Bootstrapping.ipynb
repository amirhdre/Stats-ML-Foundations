# Bootstrapping Main Ideas

Bootstrapping is a powerful statistical technique for estimating quantities about a population by averaging estimates from multiple small data samples. It is a resampling method, and in this notebook, we'll dive deep into its mechanics.

### Why Bootstrapping?
Imagine testing a new drug with only a small sample size. Redoing the experiment multiple times is expensive. Bootstrapping provides a computationally efficient alternative.

To make this process engaging, we'll explore bootstrapping concepts through visualizations and Python code.

## Bootstrapping

The main idea of bootstrapping is to randomly sample with replacement from the original dataset and create a new dataset. Once we have this new dataset, we can calculate the mean value (or any other statistic) and add it to a histogram of mean values. This process is repeated multiple times to get an estimate of the distribution of the mean values.

The steps for bootstrapping are as follows:

1. **Create a bootstrapped dataset**: Sample with replacement from the original dataset to create a new dataset of the same size.
2. **Calculate a statistic**: Calculate the statistic of interest. This could be the mean, median, standard deviation, etc.
3. **Keep track of the calculated statistic**: Keep track of the calculated statistic from each bootstrapped dataset.
4. **Repeat**: Repeat steps 1-3 a large number of times.

At the end of the bootstrapping process, we will have a distribution of the statistic of interest. This distribution gives us an idea of what might happen if we redid the experiment.

From this distribution, we can calculate the standard error of the mean value from the original dataset by calculating the standard deviation of the distribution. A 95% confidence interval can be calculated as an interval that covers 95% of the bootstrap means.

Bootstrapping is a powerful technique because it can be applied to any statistic to create a histogram of what might happen if we repeated the experiment a bunch of times. We can use that histogram to calculate stuff like standard errors and confidence intervals without having to worry about whether or not there is a nice formula.

Let's illustrate this with some Python code.


### Adding More Visualizations

Below are additional plots for:
- A sample bootstrap dataset.
- The progression of calculated means over iterations.

```python
# Visualization of one bootstrap sample
bootstrap_sample = np.random.choice(original_dataset, size=len(original_dataset), replace=True)
plt.bar(range(len(original_dataset)), bootstrap_sample, color='skyblue')
plt.title('Example of a Bootstrap Sample')
plt.xlabel('Sample Index')
plt.ylabel('Sample Value')
plt.show()
```

```python
# Visualization of means progression
progression_means = [np.mean(np.random.choice(original_dataset, size=len(original_dataset), replace=True))
                     for _ in range(50)]
plt.plot(range(1, 51), progression_means, marker='o', linestyle='--')
plt.title('Progression of Bootstrap Means (First 50 Samples)')
plt.xlabel('Iteration')
plt.ylabel('Mean Value')
plt.show()
```

```python
np.std(progression_means)
```

The above code creates a bootstrap sample from the original dataset and calculates the mean of the bootstrap sample. It repeats this process 10000 times and then plots a histogram of the bootstrap means. It also calculates the standard error and a 95% confidence interval.

From the histogram, we can see that if we redid the experiment, there's a high likelihood we will get something close to the original mean. And getting something really far from the original mean should be relatively rare. From the 95% confidence interval, we see that the interval covers zero. Therefore, we cannot reject the hypothesis that the drug is not doing anything.

Bootstrapping is a robust and flexible method that can be applied to any statistic. However, it is computationally intensive as it involves creating a large number of bootstrap samples.
