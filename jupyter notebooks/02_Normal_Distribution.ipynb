# The Normal Distribution
The normal distribution, also known as Gaussian distribution, is a probability distribution that is symmetric about the mean, showing that data near the mean are more frequent in occurrence than data far from the mean. It is often referred to as a bell-shaped curve because of its characteristic shape.

The normal distribution is incredibly important in statistics because many real-world phenomena are normally distributed. Physical quantities that are expected to be the sum of many independent processes (such as measurement errors) often have a distribution very close to the normal. Moreover, Gaussian distributions have some unique properties that are valuable in analytic studies.

## Probability Density Function
The probability density of the normal distribution is:

$$ P(x) = \frac{1}{\sqrt{2\pi\sigma^{2}}}e^{ -\frac{(x-\mu)^2}{2\sigma^{2}} } $$

Where:
- $\mu$ is the mean or expectation of the distribution (and also its median and mode)
- $\sigma$ is the standard deviation
- $\sigma^{2}$ is the variance

## Properties of a Normal Distribution
- The mean, mode and median are all equal.
- The curve is symmetric at the center (i.e. around the mean, $\mu$).
- Exactly half of the values are to the left of center and exactly half the values are to the right.
- The total area under the curve is 1.
- Approximately 68% of the data falls within one standard deviation of the mean.
- Approximately 95% of the data falls within two standard deviations of the mean.
- Approximately 99.7% of the data falls within three standard deviations of the mean.

The normal distributions occurs often in nature. For example, it describes the commonly occurring distribution of samples influenced by a large number of tiny, random disturbances, each with its own unique distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

mu, sigma = 175, 2
x = np.arange(160,190,0.1)
y = 1/(np.sqrt(2*np.pi*sigma**2))*np.exp(-(x-mu)**2/(2*sigma**2))
lower_95_bound = mu - sigma*2
higher_95_bound = mu + sigma*2
plt.figure(figsize=(10,5))
plt.plot(x, y)
plt.axvline(lower_95_bound, c='r', lw=3, alpha=0.5)
plt.axvline(higher_95_bound, c='r', lw=3, alpha=0.5)
plt.show()
```

## Example of Normal Distribution

A common example of normal distribution is human height. If we collect height data of a large group of people, we might find that the distribution of these data points is bell-shaped. The average height (mean) is at the center of the distribution, and the distribution is symmetric around the mean.

The x-axis represents the height measurements, while the y-axis represents the relative probability of observing someone of a certain height. It's relatively rare to see someone who is extremely short or extremely tall, which is why the bell-shaped curve is relatively low at the two ends. However, it's quite common to see someone who is close to the average height, so the bell-shaped curve is very tall at the center.

The width of the curve is defined by the standard deviation, which is a measure of how spread out the heights are around the mean. The larger the standard deviation, the wider the curve, and vice versa. By knowing the standard deviation, we can determine that approximately 95% of the height measurements fall within two standard deviations around the mean.

## Central Limit Theorem

The Central Limit Theorem (CLT) is a statistical theory that states that given a sufficiently large sample size from a population with a finite level of variance, the mean of all samples from the same population will be approximately equal to the mean of the population. Furthermore, all of the samples will follow an approximate normal distribution pattern, with all variances being approximately equal to the variance of the population divided by each sample's size.

The central limit theorem is the reason why many natural phenomena follow the normal distribution. This is because many processes are influenced by multiple factors, each with their own distribution, and the sum of these random variables tends towards a normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Normal Distribution Visualization
def normal_distribution_demo():
    mu = 0  # Mean
    sigma = 1  # Standard deviation
    x = np.linspace(-4, 4, 1000)  # Values for x-axis
    pdf = norm.pdf(x, mu, sigma)  # Probability density function

    plt.figure(figsize=(10, 6))
    plt.plot(x, pdf, label=r'Normal Distribution: $\mu=0$, $\sigma=1$', color='blue')
    plt.title('Normal Distribution')
    plt.xlabel('x')
    plt.ylabel('Probability Density')
    plt.grid(True)
    plt.legend()
    plt.show()

# Execute demonstrations
normal_distribution_demo()
```

```python
# Central Limit Theorem Visualization
def central_limit_theorem_demo():
    np.random.seed(42)  # For reproducibility
    num_samples = 1000  # Number of samples per experiment
    num_experiments = [1, 5, 30, 100]  # Number of experiments to average over

    plt.figure(figsize=(12, 8))
    for i, experiments in enumerate(num_experiments):
        means = [np.mean(np.random.uniform(-1, 1, num_samples)) for _ in range(experiments)]
        plt.subplot(2, 2, i + 1)
        plt.hist(means, bins=20, density=True, alpha=0.6, color='skyblue', edgecolor='black')
        mu, sigma = norm.fit(means)
        x = np.linspace(min(means), max(means), 1000)
        pdf = norm.pdf(x, mu, sigma)
        plt.plot(x, pdf, 'r-', label=f'Fit: $\mu={mu:.2f}$, $\sigma={sigma:.2f}$')
        plt.title(f'{experiments} Experiments Averaged')
        plt.xlabel('Sample Mean')
        plt.ylabel('Density')
        plt.legend()
        plt.grid(True)

    plt.tight_layout()
    plt.show()

central_limit_theorem_demo()
```
