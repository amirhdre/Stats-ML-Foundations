## Quick Review of Derivatives

Before diving into the chain rule, let's briefly review the concept of a derivative. A derivative represents the rate of change of a function with respect to one of its variables.

For example, if we have a function:

$f(x) = x^2$

The derivative $f'(x)$ with respect to $x$ is:

$f'(x) = 2x$

This tells us how the value of $f(x)$ changes as $x$ changes.

### Visual Example

Run the cell below to plot a parabola and its tangent lines to visualize derivatives.


```python

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 500)
y = x**2

# Derivative (slope) calculation
dy_dx = 2 * x

# Plot the parabola and its tangent lines
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='f(x) = x²', color='blue')
plt.plot(x, dy_dx, label="f'(x) = 2x (slope)", linestyle='--', color='orange')
plt.title("Visualization of a Parabola and its Derivative")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.grid()
plt.show()
```


## The Chain Rule

The chain rule allows us to compute the derivative of a composite function. If a variable $z$ depends on $y$, which in turn depends on $x$, then $z$ depends on $x$. The chain rule states:

$\frac{dz}{dx} = \frac{dz}{dy} \cdot \frac{dy}{dx}$

### Example: Weight, Height, and Shoe Size

Suppose:

- Weight predicts height: $h(w) = 2w$
- Height predicts shoe size: $s(h) = \frac{1}{4}h$

The chain rule helps us find how shoe size changes with weight:

$\frac{ds}{dw} = \frac{ds}{dh} \cdot \frac{dh}{dw}$

Here:

$\frac{ds}{dh} = \frac{1}{4}, \quad \frac{dh}{dw} = 2$

Thus:

$\frac{ds}{dw} = \frac{1}{4} \cdot 2 = \frac{1}{2}$


```python

# Example visualization for weight, height, and shoe size relationship
weight = np.linspace(0, 10, 50)
height = 2 * weight
shoe_size = 0.25 * height

plt.figure(figsize=(12, 8))
plt.plot(weight, height, label="Height vs. Weight (h(w) = 2w)", color='green')
plt.plot(height, shoe_size, label="Shoe Size vs. Height (s(h) = 1/4 h)", color='orange')
plt.xlabel("Weight or Height")
plt.ylabel("Height or Shoe Size")
plt.title("Weight, Height, and Shoe Size Relationships")
plt.legend()
plt.grid()
plt.show()
```
